```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
```

```python
import pandas as pd

# Load the uploaded dataset
file_path = r'C:\Users\ose06\OneDrive\바탕 화면\LLMREC\SmallTalk2Rec_Exp-crawling\crawling\movie_infos_241016.pkl'
movie_data = pd.read_pickle(file_path)
movie_data.keys(), list(movie_data.items())[:2]
```

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

# 1. 텍스트 임베딩 모델을 로드합니다.
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# 2. 영화의 장르, 주연 배우, 영화 설명 등을 결합한 텍스트 정보를 준비합니다.
movies_text_data = []  # 영화별 텍스트 정보를 담는 리스트
movie_titles = []  # 영화 제목을 담는 리스트

for movie_title, movie_info in movie_data.items():
    text_data = f"Title: {movie_title}, Genre: {movie_info.get('genre', '')}, Main Cast: {', '.join(movie_info.get('main_cast', []))}, Description: {movie_info.get('describe', '')}"
    movies_text_data.append(text_data)
    movie_titles.append(movie_title)

# 3. 텍스트 데이터를 임베딩하여 벡터화합니다.
movie_embeddings = model.encode(movies_text_data)

# 4. FAISS 인덱스를 만듭니다.
dimension = movie_embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)  # L2 거리 기반 검색 인덱스
index.add(np.array(movie_embeddings))  # 벡터들을 인덱스에 추가

# 5. 특정 영화와 유사한 영화를 검색하는 함수 작성
def search_similar_movies(query, top_k=5):
    query_embedding = model.encode([query])  # 검색 질의를 벡터화
    distances, indices = index.search(query_embedding, top_k)  # 가장 가까운 top_k개의 영화를 검색
    return [(movie_titles[idx], distances[0][i]) for i, idx in enumerate(indices[0])]

# 예시로 '기생충'과 유사한 영화 검색
search_similar_movies("Title: 기생충, Genre: Drama")
```
