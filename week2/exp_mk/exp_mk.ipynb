````python
import os
from dotenv import load_dotenv

# .env 파일의 환경 변수 로드
load_dotenv()

# 환경 변수 가져오기
openai_api_key = os.getenv("OPENAI_API_KEY")

from langchain.prompts import PromptTemplate
from langchain_openai import ChatOpenAI

# OpenAI LLM을 이용한 LangChain 구성
llm = ChatOpenAI(api_key=openai_api_key,
                 temperature=0.0, 
                 model='gpt-4o-mini')
````

````python
prompt = """당신은 영화 선호도 평가자입니다. 사용자가 특정 영화에 대해 좋은 점과 싫은 점을 이야기할 것입니다. 
이를 바탕으로 사용자의 선호도를 평가하고, 다음 형식에 맞춰 결과를 반환하세요:

JSON 객체는 다음 필드를 포함해야 합니다:
movie_title: 사용자 입력에 기반한 영화 제목.
preference_score: 사용자가 영화를 얼마나 선호하는지를 1에서 10까지의 척도로 나타낸 숫자 점수.
likes: 사용자가 영화를 좋아하는 이유.
dislikes: 사용자가 영화를 싫어하는 이유.

사용자 입력: {query}
JSON 객체:
"""

prompt_template = PromptTemplate(template=prompt, input_variables=["query"])
````

````python
# 유저한테 좋은 영화와 싫은 영화를 물어보고, 각각 그 이유를 물어봤다는 가정
````

## 사용자로부터 취향 추출하기 
> Q: 기억나는 영화와 그 영화의 좋은 점과 싫은 점을 이야기해줘

````python
import json

def parse_pref_to_json(pref):
    return json.loads(pref.replace("json", "").strip().replace("```", "")) 
````

````python
prompt = """You are a movie preference evaluator. The user will talk about what they liked and disliked about a specific movie. Based on this, evaluate the user's preferences by extracting the reasons for liking and disliking the movie by feature, and return the result in the following format:

The JSON object should include the following fields:
movie_title: The movie title based on the user's input.

likes: The reasons the user liked the movie, categorized by feature.
dislikes: The reasons the user disliked the movie, categorized by feature.

User input: {query}
JSON object:
"""

prompt_template = PromptTemplate(template=prompt, input_variables=["query"])
preference_eliction_chain = prompt_template | llm
````

````python
result_json1 = preference_eliction_chain.invoke({"query":"어바웃타임에서 아버지와 아들이 시간에 대해서 이야기하는 장면과 나중에 아들(주인공)이 더 이상 시간 여행을 하지 않고 현재에 살아간다는 점이 너무 인상깊었고, 또 로맨스적인 요소도 너무 좋았어"}).content
result_json2 = preference_eliction_chain.invoke({"query":"베테랑2에서 좋았던 점은 정해인 배우의 액션이 인상 깊었어. 굉장히 빠르게 진행되어서 좋았고 타격이 아닌 그라운드 기술이 중점이 된 액션이라 신선했어. 다만 스토리가 너무 허술해서 아쉬웠고, 특히 마지막 엔딩은 갑자기 끝난 느낌? 결말이 전혀 매력적으로 느껴지지 않았어"}).content
result_json3 = preference_eliction_chain.invoke({"query":"아메리칸 셰프 영화를 봤는데, 신나는 음악과 신나는 요리 편집이 보기에 되게 시원했어. 그리고 주인공이 아들과의 관계를 회복하는 과정도 되게 재밌었어."}).content
result_json4 = preference_eliction_chain.invoke({"query":"검사외전을 봤는데 너무 별로였어. 강동원 특유의 신나고 재밌는 분위기는 좋았지만 스토리가 너무 별로였어. 문제가 해결되는 과정이 너무 얼렁뚱땅 넘어가는 느낌이라고 할까? 전혀 설득력이 없었어."}).content
````

````python
result_json1 = parse_pref_to_json(result_json1)
result_json2 = parse_pref_to_json(result_json2)
result_json3 = parse_pref_to_json(result_json3)
result_json4 = parse_pref_to_json(result_json4)
````

````python
print(result_json1)
print(result_json2)
print(result_json3)
print(result_json4)
````

````python
dict_list = [result_json1, result_json2, result_json3, result_json4]

combined_dict = {}

for movie_dict in dict_list:
    # Using movie title as the key to avoid overwriting
    movie_title = movie_dict['movie_title']
    combined_dict[movie_title] = {
        'likes': movie_dict.get('likes', {}),
        'dislikes': movie_dict.get('dislikes', {})
    }

combined_json = json.dumps(combined_dict, ensure_ascii=False,  indent=4)
print(combined_json)
````

## 리뷰 데이터에서 영화의 좋은 점, 나쁜 점 추출하기

````python
import pandas as pd

df = pd.read_csv('./movie_infos.csv')
df = df.dropna()

import pickle

with open('../../crawling\data\movie_infos_241017.pkl', 'rb') as f:  # 'rb'는 read binary의 약자
    movie_infos = pickle.load(f)
````

````python
prompt = """You are a movie review analyst. Analyze the reviews provided for the given movie and extract the positive and negative aspects of the movie for each feature (such as plot, acting, cinematography, soundtrack, etc.) in Korean.

The result should be returned as a JSON object in the following format:

The JSON object should include the following fields:
movie_title: The title of the movie based on user input.
features: A list of features where each feature contains:
  - feature_name: The name of the feature (e.g., plot, acting, cinematography, soundtrack).
  - likes: The positive aspects of the feature extracted from the reviews in Korean.
  - dislikes: The negative aspects of the feature extracted from the reviews in Korean.

Movie title: {title}
User input (reviews): {reviews}
JSON object (in Korean):
"""

prompt_template = PromptTemplate(template=prompt, input_variables=["query"])
review_analysis_chain = prompt_template | llm
````

````python
review_jsons = []
for i in range(20):
   review_jsons.append(review_analysis_chain.invoke({"title":df.iloc[i]['영화제목'], "reviews":"\n".join(df.iloc[i]['코멘트 10개 멘트 (|| parser)'].split('||'))}).content)
````

````python
def parse_pref_to_json(pref):
    return json.loads(pref.replace("json", "").strip().replace("```", "")) 

review_json_list = [parse_pref_to_json(review_json) for review_json in review_jsons]
````

````python
for review_json in review_json_list:
    print(review_json)
````

````python
combined__review_dict = {}

for movie_dict in review_json_list:
    # Using movie title as the key to avoid overwriting
    movie_title = movie_dict['movie_title']
    combined__review_dict[movie_title] = {
        'features': movie_dict.get('features', {})
        }

combined_review_json = json.dumps(combined__review_dict, ensure_ascii=False,  indent=4)
print(combined_review_json)
````

````python
import pickle 
def save_file_to_pickle(file:str, file_path:str):
    """
    file: 저장할 파일
    file_path: 저장할 이름 (~.pkl)
    """
    # pickle 파일로 저장
    with open(file_path, 'wb') as f:  # 'wb'는 write binary의 약자
        pickle.dump(file, f)
        
def load_file_from_pickle(file_path:str):
    # pickle 파일에서 데이터 불러오기
    with open(file_path, 'rb') as f:  # 'rb'는 read binary의 약자
        file = pickle.load(f)
        
    return file
````
