```python
from langchain_community.embeddings import HuggingFaceBgeEmbeddings

embeddings = HuggingFaceBgeEmbeddings(model_name="BAAI/bge-m3")
```

```python
from langchain_chroma import Chroma

vector_store = Chroma(
    collection_name="movie_plot",
    embedding_function=embeddings,
    persist_directory="../data/chroma",  # Where to save data locally, remove if not necessary
)
```

```python
contexts = vector_store.similarity_search(
    "어린이가 좋아하는 영화",
    k=3,
    filter={
        "adult": {
            
        }
    }
)
```

```python
import pandas as pd

df = pd.read_csv("../data/movies_sample_cleaned.csv")
```

```python
df.head()
```

```python
print(df['comment_texts'].iloc[0])
```

```python
import ast

df["comment_texts"] = df["comment_texts"].apply(ast.literal_eval)
```

```python
def func(comment_texts):
    result = []
    for comment in comment_texts:
        result += [comment.replace("\n", " ")]
    return "\n\n".join(result)

df["comment_texts"] = df["comment_texts"].apply(func)
```

```python
print(df["comment_texts"].iloc[0])
```

```python
df.to_csv("../data/movies_sample_cleaned.csv", index=False)
```
